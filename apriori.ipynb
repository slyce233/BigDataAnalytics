```python
import pandas as pd
from apyori import apriori
```

```python
retail = pd.read_csv('http://fimi.uantwerpen.be/data/retail.dat')
```

```python
records = retail.values.tolist()
for i in range(len(records)):
    records[i] = records[i][0].split()
```

```python
association_rules = apriori(records, min_support=0.0045, min_confidence=0.20, min_lift=3, min_length=2)
association_results = list(association_rules)
```

```python
results = []
for item in association_results:
    pair = item[0]
    items = [x for x in pair]
    
    item0 = items[0]
    item1 = items[1]
    support = str(item[1])[:7]
    confidence = str(item[2][0][2])[:7]
    lift = str(item[2][0][3])[:7]

    rows = (item0, item1,support,confidence,lift)
    results.append(rows)
```

```python
for i in range(len(results)):
    print('====================================')
    print('ITEMS:\t\t', results[i][0], '-->',  results[i][1])
    print('SUPPORT:\t', results[i][2])
    print('CONFIDENCE:\t', results[i][3])
    print('LIFT:\t\t', results[i][4])

```
